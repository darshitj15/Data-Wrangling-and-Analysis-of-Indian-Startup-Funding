```python
def same_citynames(x):
#     & and ,
    if re.search('newdelhi',x):
        return x.replace('newdelhi','delhi')
    if re.search('nwdelhi',x):
        return x.replace('nwdelhi','delhi')
    if re.search(r'bangalore',x):
        return x.replace('bangalore','bengaluru')
#     delhi new delhi
    if re.search(r'california',x):
        return x.replace('california','unitedstates')
    if re.search(r'sanjose',x):
        return x.replace('sanjose','unitedstates')
    if re.search(r'newyork',x):
        return x.replace('newyork','unitedstates')
    if re.search(r'sanfrancisco',x):
        return x.replace('sanfrancisco','unitedstates')
    if re.search(r'burnsville',x):
        return x.replace('burnsville','unitedstates')
    if re.search(r'dallas',x):
        return x.replace('dallas','unitedstates')
    if re.search(r'us',x):
        return x.replace('us','unitedstates')
    if re.search(r'usa',x):
        return x.replace('usa','unitedstates')
    if re.search(r'sfo',x):
        return x.replace('sfo','unitedstates')
    if re.search(r'boston',x):
        return x.replace('boston','unitedstates')
    if re.search(r'missourie',x):
        return x.replace('missourie','unitedstates')
    if re.search(r'menlopark',x):
        return x.replace('menlopark','unitedstates')
    if re.search(r'paloalto',x):
        return x.replace('paloalto','unitedstates')
    if re.search(r'santamonica',x):
        return x.replace('santamonica','unitedstates')
    if re.search(r'andheri',x):
        return x.replace('andheri','mumbai')
    if re.search(r'chembur',x):
        return x.replace('chembur','mumbai')
    if re.search(r'ahemdabad',x):
        return x.replace('ahemdabad','ahmedabad')
    if re.search(r'panaji',x):
        return x.replace('panaji','goa')
    else:
        return x
def remove_and(x):
    if re.search(r'and',x):
        a=x.split('and')
        return a[0]
    else:
        return x
```
